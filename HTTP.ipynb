```
In HTTP headers:

[Field] General: include following methods~
[Method] Request URL, Request Method, Status Code, Remote Address, Referrer Policy

[Field] Response:
[Method] Server, set-Cookie, Content-Type, Content-Length, Date

[Field] Request:
[Method] Cookies, Accept-XXX, Content-Type, Content-Length, Authorization, User-Agfent, Referrer
```





```
HTTP Status Codes:
1**: Informational [Request recieved / processing]

2**: Success [Successfully Recieved, understood and accepted]

3**: Redirect [Further action must be taken/ redirect]

4**: Client Error [Request does not have what it needs]

5**: Server Error [Server failded to fulfil an apparent valid request]

--------------------------------------------------------
Exp:

200 -- ok

201 -- ok but created something

301 -- Moved to new URL

304 -- Not modified (Cached version)

400 -- Bad request

401 -- Unauthorized

404 -- Not found

500 -- Internal server error
```
