's' stand for subsitute
'g' stand for till end
'.' stand for any charactors
'*' stand for whatever numbers charactors after

replace the first element match

```python
sed 's/oldword/newword/' file
```

replace everything match in the line

```python
sed 's/oldword/newword/g' file
```

Delete every thing after something

```python
sed 's/something.*//g' # Also, we can say repalce every thing with empty 
```

-i option makes the out put back into file (change the file)

```python
sed -i 's/oldword/newword/g' file
```
