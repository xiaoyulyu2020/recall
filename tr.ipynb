```python
tr 'old' 'new'<=====all o change to n, all l change to e, all d change to w
tr ' ' '_'
tr '[:lower:]' '[:upper:]'<===change lowers to upper space
```

```python
tr -d ' '<--remove space
tr ' ' '\t' <===change space to tab
```

```python
tr -c 

#if we have a string "this is a string"

tr -d ' ' or tr -d '[:space:]' #we will have "thisisastring"

tr -dc ' ' or tr -dc '[:space:]' # we will have the "  " only space

-c stand for the opposite way
so -dc means 'we remove everything but ()'
```
